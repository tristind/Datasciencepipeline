```python
# ============================== loading libraries ===========================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.cross_validation import cross_val_score
from collections import Counter


# loading training data
#Data munging
df = pd.read_table('pokemon_alopez247.csv',sep=",")
df['Type_1'].replace('Bug',1,inplace=True)
df['Type_1'].replace('Dark',2,inplace=True)
df['Type_1'].replace('Dragon',3,inplace=True)
df['Type_1'].replace('Electric',4,inplace=True)
df['Type_1'].replace('Fairy',5,inplace=True)
df['Type_1'].replace('Fighting',6,inplace=True)
df['Type_1'].replace('Fire',7,inplace=True)
df['Type_1'].replace('Flying',8,inplace=True)
df['Type_1'].replace('Ghost',9,inplace=True)
df['Type_1'].replace('Grass',10,inplace=True)
df['Type_1'].replace('Ground',11,inplace=True)
df['Type_1'].replace('Ice',12,inplace=True)
df['Type_1'].replace('Normal',13,inplace=True)
df['Type_1'].replace('Poison',14,inplace=True)
df['Type_1'].replace('Psychic',15,inplace=True)
df['Type_1'].replace('Rock',16,inplace=True)
df['Type_1'].replace('Steel',17,inplace=True)
df['Type_1'].replace('Water',18,inplace=True)

 
#Machine Learning
# create design matrix X and target vector y
X = np.array(df.ix[:, 5:9]) 	# end index is exclusive
y = np.array(df['Type_1']) 	# showing you two ways of indexing a pandas df

# split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
# ============================== KNN with k = 1 ===============================================
# instantiate learning model (k = 1)
knn = KNeighborsClassifier(n_neighbors=1)

# fitting the model
knn.fit(X_train, y_train)

# predict the response
pred = knn.predict(X_test)

# evaluate accuracy
acc = accuracy_score(y_test, pred) * 100
print('\nThe accuracy of the knn classifier for k = 1 is %d%%' % acc)
# ============================== parameter tuning =============================================
# creating odd list of K for KNN
myList = list(range(0,50))
neighbors = list(filter(lambda x: x % 2 != 0, myList))

# empty list that will hold cv scores
cv_scores = []

# perform 10-fold cross validation
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())
    print(k, scores.mean())

# changing to misclassification error
MSE = [1 - x for x in cv_scores]

# determining best k
optimal_k = neighbors[MSE.index(min(MSE))]
print('\nThe optimal number of neighbors is %d.' % optimal_k)

# plot misclassification error vs k 
plt.plot(neighbors, MSE)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Misclassification Error')
plt.show()
```
